```python
import pandas as pd

url = 'https://www.infomoney.com.br/cotacoes/empresas-b3/'
tables = pd.read_html(url)

# Verifica quantas tabelas existem
print(f"Número de tabelas encontradas: {len(tables)}")

# Mostra as primeiras tabelas
for i, tabela in enumerate(tables):
    print(f"\nTabela {i}:\n")
    print(tabela.head())
    print(tabela.columns)
```
